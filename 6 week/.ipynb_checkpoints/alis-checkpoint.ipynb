Add features based on the session start time: hour, whether it's morning, day or night and so on.

```python
# You code here

def time_to_rad (t):
    hour = datetime.strptime(t, '%Y-%m-%d %H:%M:%S').hour
    minute = datetime.strptime(t, '%Y-%m-%d %H:%M:%S').minute
    second = datetime.strptime(t, '%Y-%m-%d %H:%M:%S').second
    seconds_in_day = 86400
    rad = (hour * 3600 + minute * 60 + second) / seconds_in_day * 2 * math.pi
    return rad

def add_time_start_column (dataframe):
    df = dataframe.copy()
    df['time_rad'] = df['time1'].map(lambda x: time_to_rad(x)) 
    df['time_start_x'] = df['time_rad'].map(lambda x: math.cos(x)) 
    df['time_start_y'] = df['time_rad'].map(lambda x: math.sin(x))
    df.drop('time_rad', axis=1, inplace=True)
    
    return df

#for index, row in df.iterrows():
#    row['site1']

df_sample = add_time_start_column (df_sample)     
df_sample.head()
```

```python
def date_to_rad (time):
    year = datetime.strptime(time, '%Y-%m-%d %H:%M:%S').year
    date = datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
    first_day = datetime.strptime(str(year) + "-1-1", "%Y-%m-%d")
    day_number = (date - first_day).days
    rad = day_number / 365 * 2 * math.pi
    return rad

def add_date_start_column (dataframe):
    df = dataframe.copy()        
    df['date_rad'] = df['time1'].map(lambda x: date_to_rad(x)) 
    df['date_start_x'] = df['date_rad'].map(lambda x: math.cos(x)) 
    df['date_start_y'] = df['date_rad'].map(lambda x: math.sin(x))
    df.drop('date_rad', axis=1, inplace=True)
    return df

df_sample = add_date_start_column (df_sample)     
df_sample.head()
```
